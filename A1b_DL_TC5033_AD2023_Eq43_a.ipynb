```python
#Loop
def pix_mapped(list_files_pix):
  data_out=[]
  for i in range(len(list_files_pix)):
    img_path = list_files_pix[i][2]
    img = Image.open(img_path)
    img = img.resize((50,50))
    img = img.convert('LA')
    img_ar = np.array(img)
    data_out.append(list_files_pix[i])
    for j in range(len(img_ar[0])):
      for k in range(len(img_ar[0][0])):
        data_out[len(data_out)-1].append(img_ar[j][k][0])
    if i % 100 == 0: CSV_Out(data_out,"FOut_Img_"+str(i)); data_out=[];
  return list_files_pix
```

```python
def CSV_Out(data,file_name):
  np.savetxt(file_name+".csv", data, delimiter=',', fmt='%s', header='Name,Value,Quantity', comments='')
```
